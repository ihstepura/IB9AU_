```python
#Task 2
#Ihor Stepura
#Insights: Autograd automatically applies the chain rule across a computation graph to produce correct partial derivatives for a function that mixes nonlinear activations (sigmoid), exponentiation, multiplication, and division: by setting requires_grad=True on inputs,
#computing a scalar output z, and calling z.backward(), you get variables stored in w.grad, x.grad, and b.grad without doing any manual calculus.
#The key takeaway is that Autograd tracks every differentiable operation you perform on tensors, builds the graph dynamically,
#and then computes gradients in reverse (backprop), making it easy to verify analytical derivatives and extend to much more complex models with the same workflow.
import torch

# Given Initial Values
w_val = 2.0
x_val = 4.0
b_val = 1.5

# 1. Create three tensors for w, x, and b with the given initial values.
#    Make sure PyTorch tracks their gradients by setting requires_grad=True.
w = torch.tensor(w_val, requires_grad=True)
x = torch.tensor(x_val, requires_grad=True)
b = torch.tensor(b_val, requires_grad=True)

print(f"Initial w: {w}")
print(f"Initial x: {x}")
print(f"Initial b: {b}")

# 2. Write the Python code to compute z using these tensors.
#    z = σ(w * x^2) + 1 / b^3
#    where σ is the sigmoid function, which in PyTorch is torch.sigmoid().

sigmoid_input = w * (x**2)
sigma_term = torch.sigmoid(sigmoid_input)

power_term = 1 / (b**3)

z = sigma_term + power_term

print(f"\nComputed z: {z}")

# 3. Call the .backward() method to compute the gradients.
z.backward()

# 4. Print the computed gradients for w, x, and b.
print(f"\nGradient of z with respect to w (dz/dw): {w.grad}")
print(f"Gradient of z with respect to x (dz/dx): {x.grad}")
print(f"Gradient of z with respect to b (dz/db): {b.grad}")
```
